```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Filtering data for Ital and non zero cases  
df = pd.read_csv('covid-data.csv')
df = df[df['location'] == 'Italy']
df = df[df['total_cases'] > 0]
```

```python
# Selecting desired columns
df = df.filter(items=['date', 'total_cases','new_cases', 'total_deaths'])
cols = df.columns.drop('date')
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df = df.set_index('date')
df = df.loc['2020-03-09':]
```

```python
# plotting from 9th March
df.plot(colormap='Set1', figsize=(12,5), lw=2);
plt.legend(["Toal covid-19 cases", "New covid-19 cases", "Total deaths"], frameon=False);
plt.xlabel('Date [yyyy-mm-dd]');
plt.ylabel('Number')
plt.title('Covid-19 cases and related fatalities in Italy since lockdown')
plt.savefig('covid-19.png')
plt.show()
```
